```python
import json, requests, numpy as np, pandas as pd
```

```python
URL = 'https://fantasy.premierleague.com/api/bootstrap-static/'
r = requests.get(URL)
data = r.json()
```

```python
elements_df = pd.DataFrame(data['elements'])
elements_types_df = pd.DataFrame(data['element_types'])
```

```python
sel_elements = elements_df[['second_name', 'total_points','element_type',"points_per_game", "selected_by_percent", "goals_scored", "assists", "bonus","influence","creativity","threat","ict_index","expected_goals","expected_assists","expected_goal_involvements","expected_goals_per_90","expected_assists_per_90","expected_goal_involvements_per_90"]]
```

```python
sel_elements['position'] = sel_elements.element_type.map(elements_types_df.set_index('id').singular_name)
```

```python
final = sel_elements.drop(['element_type'], axis=1)
```
